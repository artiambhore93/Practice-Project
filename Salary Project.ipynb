```python
print('max:',df['salary'].max())
print('min:',df['salary'].min())
add=df['salary'].max()-df['salary'].min()
grp=round(add/3,3)
print('range difference:',(grp))
```

```python
low=df['salary'].min()+grp
mid=low+grp

print('upper bound of Low grp',low)
print('upper bound of Mid grp',mid)
print('upper bound of High grp','max:',df['salary'].max())
```
